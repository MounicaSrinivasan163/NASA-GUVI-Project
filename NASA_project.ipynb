```python
asteroids_data = []
target = 10000
url  = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-08&api_key={api_key}"

while len(asteroids_data) < target:
  response = requests.get(url)
  data = response.json()
  details = data['near_earth_objects']
  for date, asteroid_list in details.items():
      for asteroid in asteroid_list:
          asteroids_data.append(dict(id = int(asteroid['id']),
                                     neo_reference_id = int(asteroid['neo_reference_id']),
                                     name = asteroid['name'],
                                     absolute_magnitude_h = asteroid['absolute_magnitude_h'],
                                     estimated_diameter_min_km = asteroid['estimated_diameter']['kilometers']['estimated_diameter_min'],
                                     estimated_diameter_max_km = asteroid['estimated_diameter']['kilometers']['estimated_diameter_max'],
                                     ispotentially_hazardous_asteroid = asteroid['is_potentially_hazardous_asteroid'],
                                     close_approach_date = datetime.strptime(asteroid['close_approach_data'][0]['close_approach_date'], "%Y-%m-%d").date(),
                                     relative_velocity_kmph = float(asteroid['close_approach_data'][0]['relative_velocity']['kilometers_per_hour']),
                                     astronomical = float(asteroid['close_approach_data'][0]['miss_distance']['astronomical']),
                                     orbiting_body = asteroid['close_approach_data'][0]['orbiting_body'],
                                     miss_distance_km = float(asteroid['close_approach_data'][0]['miss_distance']['kilometers']),
                                     miss_distance_lunar = float(asteroid['close_approach_data'][0]['miss_distance']['lunar'])))

          if len(asteroids_data) >= target:
              break
      if len(asteroids_data) >= target:
              break
  url = data['links']['next']
```

# **MYSQL CONNECTION**

```python
import mysql.connector

con_obj = mysql.connector.connect(
    host='localhost',
    user='root',
    password='root',
    port=3306
)
print("Connected!")
```

```python
cursor=con_obj.cursor()
```

```python
cursor.execute("create database nasa_asteroid_db;") #creates database

# Show available databases
cursor.execute("show databases;")
for db in cursor:
    print(db)
```

```python
# Switch to the database 
cursor.execute("USE nasa_asteroid_db;")

# Create asteroids table
cursor.execute('''
    CREATE TABLE asteroids (
        id INT,
        name VARCHAR(100),
        absolute_magnitude_h FLOAT,
        estimated_diameter_min_km FLOAT,
        estimated_diameter_max_km FLOAT,
        is_potentially_hazardous_asteroid INT
    )
''')

# Create close_approach table
cursor.execute('''
    CREATE TABLE close_approach (
        neo_reference_id INT,
        close_approach_date DATE,
        relative_velocity_kmph FLOAT,
        astronomical FLOAT,
        miss_distance_km FLOAT,
        miss_distance_lunar FLOAT,
        orbiting_body VARCHAR(100)
         )
''')

# Show all tables
cursor.execute("SHOW TABLES")
for table in cursor:
    print(table)

```

```python
for _, row in df.iterrows():
    cursor.execute('''
        INSERT INTO asteroids (
            id, name, absolute_magnitude_h,
            estimated_diameter_min_km, estimated_diameter_max_km,
            is_potentially_hazardous_asteroid
        )
        VALUES (%s, %s, %s, %s, %s, %s)
    ''', (
        int(row['id']),
        row['name'],
        float(row['absolute_magnitude_h']),
        float(row['estimated_diameter_min_km']),
        float(row['estimated_diameter_max_km']),
        int(row['ispotentially_hazardous_asteroid'])
    ))
```

```python
for _, row in df.iterrows():
    cursor.execute('''
        INSERT INTO close_approach (
            neo_reference_id, close_approach_date, relative_velocity_kmph,
            astronomical, miss_distance_km, miss_distance_lunar, orbiting_body
        )
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    ''', (
        int(row['neo_reference_id']),
        row['close_approach_date'],  # Can be datetime.date or string in 'YYYY-MM-DD'
        float(row['relative_velocity_kmph']),
        float(row['astronomical']),
        float(row['miss_distance_km']),
        float(row['miss_distance_lunar']),
        row['orbiting_body']
    ))
```
